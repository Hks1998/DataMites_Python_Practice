```python
#Student Grade System
name=input("Enter student name:")
ma=int(input("Enter marks in Maths:"))
sci=int(input("Enter marks in Science:"))
eng=int(input("Enter marks in English:"))
sst=int(input("Enter marks in Social Science:"))
hin=int(input("Enter marks in Hindi:"))
total=ma+sci+eng+sst+hin
percentage=(total/500)*100
if percentage>=90:
    grade="A"
elif percentage>=80:
    grade="B"
elif percentage>=70:
    grade="C"
elif percentage>=60:
    grade="D"
else:
    grade="F"
print("Marksheet:")
print("Name:", name)
print("Marks in Maths:", ma)
print("Marks in Science:", sci)
print("Marks in English:", eng)
print("Marks in Social Science:", sst)
print("Marks in Hindi:", hin)
print("Total Marks:", total)
print(f"Percentage: {percentage:.2f}%")
print("Grade:", grade)
```

```python
#Password Validator

# Password Validator

password = input("Enter your password: ")

special_chars = "@#_$%^&*!"

if len(password) < 8:
    print("Password is too short! It must contain at least 8 characters.")
else:
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
        elif ch in special_chars:
            has_special = True

    if has_upper and has_lower and has_digit and has_special:
        print("Password is valid.")
    else:
        print("Password is invalid.")
        if not has_upper:
            print("- Password must contain at least one uppercase letter.")
        if not has_lower:
            print("- Password must contain at least one lowercase letter.")
        if not has_digit:
            print("- Password must contain at least one digit.")
        if not has_special:
            print("- Password must contain at least one special character.")
```

```python
#Bank login system

import json

with open("users.json", "r") as file:
    user_data = json.load(file)

attempts = 3
login = False

while attempts > 0 and not login:
    uname = input("Enter Username: ")
    pwd = input("Enter Password: ")

    found = False

    for user in user_data:
        if uname == user["userName"]:
            found = True

            if pwd == user["password"]:
                print("Login Successful!!")
                print("Welcome", uname)
                login = True
                break
            else:
                attempts -= 1
                print("Invalid Password!")
                print("Attempts remaining:", attempts)
                break

    if not found:
        attempts -= 1
        print("Invalid Username!")
        print("Attempts remaining:", attempts)

if not login:
    print("Account Locked!")

```

```python

#Shopping Cart
import json
with open("products.json", "r") as file:
    p_data = json.load(file)


prod=[]

while True:
    print("Welcome to ABC Mart")
    print("Menu")
    print("1.View Products")
    print("2.Shopping Menu")
    print("3.Exit")
    ch=input("Enter your choice(1/2/3):")
    if ch=="1":
        print("Available Items:")
        for i in range (1,len(p_data)):
            print(f"{i}",p_data[i])
            print()
    elif ch=="2":

        print("Shopping Menu:")
        print("1.View cart")
        print("2.Add item to cart")
        print("3.Remove item from cart")
        print("4.Checkout")
        cc=input("Enter your choice(1-5):")
        if cc=="1":
            print("Your Shopping Cart:")
            for item in prod:
                print(item)
        elif cc=="2":
            print("Available Items:")
            for i in range (1,len(p_data)):
                print(f"{i}",p_data[i])
            print()
            print("Menu:")
            print("Add to Cart")
            cm=input("Press 1 to add to cart :")
            if cm=="1":
                ino=int(input("Enter product number to add to cart "))
                prod.append(p_data[ino])
            print("Product Added:",prod)
        elif cc=="3":
            print ("Items in cart:")
            for j in range (0,len(prod)):
                print(prod[j])
            irm=int(input("Enter product number to remove from cart "))
            prod.remove(prod[irm])
            print("Product Removed:",prod[irm])
        elif cc=="4":
            i_cart=[product["ProductName"] for product in prod]
            prices = [int(product["Price"]) for product in prod]
            print("Items in your cart: ")
            print(i_cart)
            total=sum(prices)
            print("Amount Payable:Rs.",total,"/-")
            allow=input("Type Y to checkout or N to revisit cart:")
            if allow=="Y":
                print("Checkout complete!!!! Your order will be delivered shortly")
                print("Thank you for shopping with US")
                break
            elif allow=='N':
                 print("Items in your cart: ")
                 print(i_cart)

    if ch=="3":
        print("Thank you for shopping Please visit again!!!")
        break
```

```python
#Employee Management System
import json


while True:
    print("Welcome to the Employee Management System for Hks Inc.")
    print("Menu")
    print("1.Add Employee")
    print("2.View all Employees")
    print("3.Search by ID")
    print("4.Exit")
    ch=input("Enter your choice(1/2/3):")
    if ch=="1":
        EmpID=int(input("Enter employee ID:"))
        Name=input("Enter Name:")
        Department=input("Enter Department")

        employee = {
            "EmpID": EmpID,
            "Name": Name,
            "Department": Department
        }
        with open("employee.json", "r") as file:
            emp_data = json.load(file)

        emp_data.append(employee)

        with open("employee.json", "w") as file:
            json.dump(emp_data, file, indent=4)
        print("Employee added successfully!")

    elif ch == "2":
        with open("employee.json", "r") as file:
            emp_data = json.load(file)

        if len(emp_data) == 0:
            print("No employees found.")
        else:
            for emp in emp_data:
                print(f"Employee ID : {emp['EmpID']}")
                print(f"Name        : {emp['Name']}")
                print(f"Department  : {emp['Department']}")
                print("-" * 30)

    elif ch == "3":
        search_id = int(input("Enter Employee ID to search: "))

        with open("employee.json", "r") as file:
            emp_data = json.load(file)

        found = False
        for emp in emp_data:
            if emp["EmpID"] == search_id:
                print("\nEmployee Found")
                print(f"Employee ID : {emp['EmpID']}")
                print(f"Name        : {emp['Name']}")
                print(f"Department  : {emp['Department']}")
                print("-" * 30)
                found = True
                break

        if not found:
            print("Employee not found.")

    elif ch == "4":
        print("Thank you!")
        break

    else:
        print("Invalid choice. Please try again.")
        
    
    



```

```python
#Student Database
import json

while True:
    print("Welcome to the Student Database Management System")
    print("Menu")
    print("1.Add Student")
    print("2.View all Students")
    print("3.Calculate total marks and percentage of student")
    print("4.Find Highest Scorer")
    print("5.Exit")
    ch=input("Enter your choice(1/2/3/4/5):")
    if ch=="1":
        StudentID=int(input("Enter student ID:"))
        Name=input("Enter Name:")
        Maths=int(input("Enter marks in Maths:"))
        Science=int(input("Enter marks in Science:"))
        English=int(input("Enter marks in English:"))
        SocialScience=int(input("Enter marks in Social Science:"))
        Hindi=int(input("Enter marks in Hindi:"))

        student = {
            "StudentID": StudentID,
            "Name": Name,
            "Maths": Maths,
            "Science": Science,
            "English": English,
            "SocialScience": SocialScience,
            "Hindi": Hindi
        }
        with open("student.json", "r") as file:
            stu_data = json.load(file)

        stu_data.append(student)

        with open("student.json", "w") as file:
            json.dump(stu_data, file, indent=4)
        print("Student added successfully!")
    
    elif ch == "2":
        with open("student.json", "r") as file:
            stu_data = json.load(file)

        if len(stu_data) == 0:
            print("No students found.")
        else:
            for stu in stu_data:
                print(f"Student ID : {stu['StudentID']}")
                print(f"Name       : {stu['Name']}")
                print(f"Maths      : {stu['Maths']}")
                print(f"Science    : {stu['Science']}")
                print(f"English    : {stu['English']}")
                print(f"Social Sci.: {stu['SocialScience']}")
                print(f"Hindi      : {stu['Hindi']}")
                print("-" * 30)
    elif ch == "3":
        search_id = int(input("Enter Student ID to calculate total marks and percentage: "))

        with open("student.json", "r") as file:
            stu_data = json.load(file)

        found = False
        for stu in stu_data:
            if stu["StudentID"] == search_id:
                total_marks = stu["Maths"] + stu["Science"] + stu["English"] + stu["SocialScience"] + stu["Hindi"]
                percentage = (total_marks / 500) * 100
                print(f"\nTotal Marks for {stu['Name']} (ID: {stu['StudentID']}): {total_marks}")
                print(f"Percentage: {percentage:.2f}%")
                found = True
                break

        if not found:
            print("Student not found.")
    elif ch == "4":
        with open("student.json", "r") as file:
            stu_data = json.load(file)

        if len(stu_data) == 0:
            print("No students found.")
        else:
            highest_scorer = None
            highest_total = -1

            for stu in stu_data:
                total_marks = stu["Maths"] + stu["Science"] + stu["English"] + stu["SocialScience"] + stu["Hindi"]
                if total_marks > highest_total:
                    highest_total = total_marks
                    highest_scorer = stu

            print(f"\nHighest Scorer: {highest_scorer['Name']} (ID: {highest_scorer['StudentID']})")
            print(f"Total Marks: {highest_total}")
        
    elif ch == "5":
        print("Thank you!")
        break
    

```
