```python
#Password Validator

# Password Validator

password = input("Enter your password: ")

special_chars = "@#_$%^&*!"

if len(password) < 8:
    print("Password is too short! It must contain at least 8 characters.")
else:
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
        elif ch in special_chars:
            has_special = True

    if has_upper and has_lower and has_digit and has_special:
        print("Password is valid.")
    else:
        print("Password is invalid.")
        if not has_upper:
            print("- Password must contain at least one uppercase letter.")
        if not has_lower:
            print("- Password must contain at least one lowercase letter.")
        if not has_digit:
            print("- Password must contain at least one digit.")
        if not has_special:
            print("- Password must contain at least one special character.")
```

```python
#Bank login system

import json

with open("users.json", "r") as file:
    user_data = json.load(file)

attempts = 3
login = False

while attempts > 0 and not login:
    uname = input("Enter Username: ")
    pwd = input("Enter Password: ")

    found = False

    for user in user_data:
        if uname == user["userName"]:
            found = True

            if pwd == user["password"]:
                print("Login Successful!!")
                print("Welcome", uname)
                login = True
                break
            else:
                attempts -= 1
                print("Invalid Password!")
                print("Attempts remaining:", attempts)
                break

    if not found:
        attempts -= 1
        print("Invalid Username!")
        print("Attempts remaining:", attempts)

if not login:
    print("Account Locked!")

```

```python
#Student Management System with OOPs

class Student:
    def __init__(self,name,age,roll_no,mqt,eng,sci,sst,hin):
        self.name=name
        self.age=age
        self.roll_no=roll_no
        self.mqt=mqt
        self.eng=eng
        self.sci=sci
        self.sst=sst
        self.hin=hin

    def get_details(self):
        name=input("Enter student name:")
        age=int(input("Enter student age:"))
        roll_no=int(input("Enter student roll number:"))
        mat=int(input("Enter marks in Maths:"))
        eng=int(input("Enter marks in English:"))
        sci=int(input("Enter marks in Science:"))
        sst=int(input("Enter marks in Social Science:"))
        hin=int(input("Enter marks in Hindi:"))
        return name,age,roll_no,mat,eng,sci,sst,hin
    def calculate_total(self):
        total=self.mqt+self.eng+self.sci+self.sst+self.hin
        return total    
    def calculate_percentage(self):
        total=self.calculate_total()
        percentage=(total/500)*100
        return percentage
    def calculate_grade(self):
        percentage=self.calculate_percentage()
        if percentage>=90:
            grade="A"
        elif percentage>=80:
            grade="B"
        elif percentage>=70:
            grade="C"
        elif percentage>=60:
            grade="D"
        else:
            grade="F"
        return grade
    def display_details(self):
        total=self.calculate_total()
        percentage=self.calculate_percentage()
        grade=self.calculate_grade()
        print("Student Details:")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Roll Number:", self.roll_no)
        print("Marks in Maths:", self.mqt)
        print("Marks in English:", self.eng)
        print("Marks in Science:", self.sci)
        print("Marks in Social Science:", self.sst)
        print("Marks in Hindi:", self.hin)
        print("Total Marks:", total)
        print(f"Percentage: {percentage:.2f}%")
        print(f"Grade: {grade}")
student1 = Student("",0,0,0,0,0,0,0)
name,age,roll_no,mat,eng,sci,sst,hin=student1.get_details()
student1=Student(name,age,roll_no,mat,eng,sci,sst,hin)
student1.display_details()  
```

```python
#employee salary system using inheritance and method overriding

class Employee: 
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def get_details(self):
        return f"Name: {self.name}, Salary: {self.salary}"

    def calculate_bonus(self):
        return 0

class Manager(Employee):
    def __init__(self, name, salary, team_size):
        super().__init__(name, salary)
        self.team_size = team_size

    def get_details(self):
        return f"Name: {self.name}, Salary: {self.salary}, Team Size: {self.team_size}, Bonus: {self.calculate_bonus()}"

    def calculate_bonus(self):
        return self.salary * 0.1 * self.team_size

class Developer(Employee):
    def __init__(self, name, salary, programming_languages):
        super().__init__(name, salary)
        self.programming_languages = programming_languages

    def calculate_bonus(self):
        return self.salary * 0.05 * len(self.programming_languages)
    def get_details(self):
        return f"Name: {self.name}, Salary: {self.salary}, Programming Languages: {', '.join(self.programming_languages)},Bonus: {self.calculate_bonus()}"
    
name = input("Enter employee name: ")
salary = float(input("Enter employee salary: "))
role = input("Enter employee role (Manager/Developer): ")
programming_languages = []
team_size = 0
if role == "Manager":   
    team_size = int(input("Enter team size: "))
    employee = Manager(name, salary, team_size)
elif role == "Developer":
    num_languages = int(input("Enter number of programming languages: "))
    for i in range(num_languages):
        language = input(f"Enter programming language {i + 1}: ")
        programming_languages.append(language)
        employee = Developer(name, salary, programming_languages)
else:
    print("Invalid role! Please enter either 'Manager' or 'Developer'.")
    exit()

print(employee.get_details())
```

```python
# Library Management System using OOP

class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrowed = False

    def get_details(self):
        status = "Borrowed" if self.is_borrowed else "Available"
        return f"Title: {self.title}, Author: {self.author}, ISBN: {self.isbn}, Status: {status}"


class Users:
    def __init__(self, name, user_id):
        self.name = name
        self.user_id = user_id
        self.borrowed_books = []

    def borrow_book(self, book):
        self.borrowed_books.append(book)

    def return_book(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)


class Library:
    def __init__(self):
        self.books = []
        self.users = []

    def add_book(self, book):
        self.books.append(book)

    def add_user(self, user):
        self.users.append(user)

    def display_books(self):
        if not self.books:
            print("No books available.")
        else:
            print("\nAvailable Books")
            print("-" * 70)
            for book in self.books:
                print(book.get_details())

    def display_users(self):
        if not self.users:
            print("No users found.")
        else:
            print("\nUser Details")
            print("-" * 70)
            for user in self.users:
                borrowed = ", ".join(book.title for book in user.borrowed_books)
                if borrowed == "":
                    borrowed = "None"

                print(f"Name: {user.name}")
                print(f"User ID: {user.user_id}")
                print(f"Borrowed Books: {borrowed}")
                print("-" * 70)

    def find_book_by_isbn(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                return book
        return None

    def find_user_by_id(self, user_id):
        for user in self.users:
            if user.user_id == user_id:
                return user
        return None



library = Library()

library.add_book(Book("The Great Gatsby", "F. Scott Fitzgerald", "9780743273565"))
library.add_book(Book("To Kill a Mockingbird", "Harper Lee", "9780061120084"))
library.add_book(Book("1984", "George Orwell", "9780451524935"))
library.add_book(Book("Pride and Prejudice", "Jane Austen", "9780141439518"))

while True:

    print("\n====== Library Management System ======")
    print("1. Add User")
    print("2. Display Books")
    print("3. Borrow Book")
    print("4. Return Book")
    print("5. Display Users")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":

        name = input("Enter User Name: ")
        user_id = input("Enter User ID: ")

        if library.find_user_by_id(user_id):
            print("User ID already exists.")
        else:
            library.add_user(Users(name, user_id))
            print("User added successfully.")

    elif choice == "2":

        library.display_books()

    elif choice == "3":

        user_id = input("Enter User ID: ")
        isbn = input("Enter Book ISBN: ")

        user = library.find_user_by_id(user_id)
        book = library.find_book_by_isbn(isbn)

        if user and book:

            if book.is_borrowed:
                print("Book is already borrowed.")

            else:
                user.borrow_book(book)
                book.is_borrowed = True
                print(f"{user.name} borrowed '{book.title}' successfully.")

        else:
            print("Invalid User ID or ISBN.")

    elif choice == "4":

        user_id = input("Enter User ID: ")
        isbn = input("Enter Book ISBN: ")

        user = library.find_user_by_id(user_id)
        book = library.find_book_by_isbn(isbn)

        if user and book:

            if book in user.borrowed_books:
                user.return_book(book)
                book.is_borrowed = False
                print(f"{user.name} returned '{book.title}' successfully.")
            else:
                print("This user has not borrowed this book.")

        else:
            print("Invalid User ID or ISBN.")

    elif choice == "5":

        library.display_users()

    elif choice == "6":

        print("Thank you for using the Library Management System!")
        break

    else:
        print("Invalid choice! Please try again.")
```

```python
#Contact Management System using OOPS


class Contact:
    def __init__(self, name, phone, email):
        self.name = name
        self.phone = phone
        self.email = email

    def get_details(self):
        return f"Name: {self.name}, Phone: {self.phone}, Email: {self.email}"
    
    def add_contact(self, name, phone, email):
        self.name = name
        self.phone = phone
        self.email = email
    def update_contact(self, name=None, phone=None, email=None):
        if name:
            self.name = name
        if phone:
            self.phone = phone
        if email:
            self.email = email
    def delete_contact(self):
        self.name = None
        self.phone = None
        self.email = None
    def display_contact(self):
        if self.name and self.phone and self.email:
            print(self.get_details())
        else:
            print("Contact has been deleted or is empty.")
    def find_contact_by_name(self, name):
        if self.name == name:
            return self
        return None
    
    def find_contact_by_phone(self, phone):
        if self.phone == phone:
            return self
        return None
    
    def find_contact_by_email(self, email):
        if self.email == email:
            return self
        return None
    
Contacts = []

while True:
    print("Welcome to the Contact Management System")
    print("Menu")
    print("1.Add Contact")
    print("2.View all Contacts")
    print("3.Search by Name")
    print("4.Search by Phone")
    print("5.Search by Email")
    print("6.Exit")
    ch=input("Enter your choice(1/2/3/4/5/6):")
    if ch=="1": 
        name=input("Enter contact name:")
        phone=input("Enter contact phone:")
        email=input("Enter contact email:")
        contact = Contact(name, phone, email)
        Contacts.append(contact)
        print("Contact added successfully!")    
        
    elif ch=="2":
        if len(Contacts) == 0:
            print("No contacts found.")
        else:
            for contact in Contacts:
                contact.display_contact()
    elif ch=="3":
        search_name = input("Enter contact name to search: ")
        found = False
        for contact in Contacts:
            if contact.find_contact_by_name(search_name):
                print("Contact Found:")
                contact.display_contact()
                found = True
                break
            if not found:
                print("Contact not found.")
    elif ch=="4":
        search_phone = input("Enter contact phone to search: ")
        found = False
        for contact in Contacts:
            if contact.find_contact_by_phone(search_phone):
                print("Contact Found:")
                contact.display_contact()
                found = True
                break
            if not found:
                print("Contact not found.")


    elif ch=="5":
        search_email = input("Enter contact email to search: ")
        found = False
        for contact in Contacts:
            if contact.find_contact_by_email(search_email):
                print("Contact Found:")
                contact.display_contact()
                found = True
                break
            if not found:
                print("Contact not found.")

    elif ch=="6":
        print("Thank you for using the Contact Management System!")
        break

```
