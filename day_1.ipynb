```python
#OOP Methods
#inheritence
#polymorphism
#encapsulation
#abstraction

class Multi():
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def Mul(self):
        return f"Product of x and y is {self.x*self.y}"
    
    
```

```python
obj=Multi(5,12)
obj.Mul()
```

```python
class Cal(Multi):
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def Add(self):
        return f"Sum of x and y is {self.x+self.y}"
    


    
```

```python
obj=Cal(5,5)
obj.Add()
```

```python
class GF:
    def Grand(self):
        print("I have Grandfather genes")

class P:
    def Parent(self):
        print("I have Parent Genes")

class Child(GF,P):
    pass



obj=Child()
obj.Grand()
obj.Parent()
```
